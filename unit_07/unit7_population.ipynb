```python
import csv
f = open('age.csv')
data = csv.reader(f)

result = []

for row in data :
    if '서현1' in row[0] : # 신도림 동을 찾고
        for i in row[3:] : # 그 행의 4번째 부터가 0세, 1세, ..., 100세이상 이므로
            result.append(int(i)) # 차트에 그리려면 int형으로
print(result)

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') # 격자무늬 흰색
plt.figure(figsize=(10, 3))
plt.title('2020 04 Population')
plt.xlabel('age')
plt.ylabel('cnt')
plt.plot(result)
plt.show()
```

```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []

name = input('인구구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요 : ')

for row in data :
    if name in row[0] : # 입력받은 지역의 행을 뽑는다
        for i in row[3:] :
            result.append(int(i.replace(',',''))) # 통화단위로 되어있음(54,322) comma 제거

plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic') # name이 한글이므로 한글처리

plt.title('04_2020_Population_' + name)
plt.xlabel('나이')
plt.ylabel('명')
plt.plot(result)
plt.show()
```
