#### bar graph
matplotlib.pyplot.bar(x, height, width, bottom, ...)
* bottom(default = 0) : y축이 시작하는 값

```python
import matplotlib.pyplot as plt
plt.bar([0, 1, 2, 4, 6, 10], [1, 2, 3, 5, 6, 7]) # x, y 값(요소갯수 같아야함)
plt.show()
```

```python
# 항아리 모양 그래프 그리기(pyramid chart)
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data :
    if '신도림' in row[0] :
        for i in range(0, 101) :
            m.append(-int(row[i+3])) # 항아리 모양으로 나타내기 위해 남자 인구수에 마이너스 붙인다
            f.append(int(row[-(i+1)]))
f.reverse() # 여자는 뒤에서부터 하므로 100세이상부터 값이 들어간다.

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # graph x축 -가 안깨지게 하기위해
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show() # m과 f가 겹쳐져서 나온다(둘다 나이는 양수이므로)
```

```python
# 입력한 동네 population
# 남_녀 분리하여 pyramid chart 그리기'

import csv

f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

name = input('읍면동 단위로 입력하세요 : ')

for row in data :
    if name in row[0] :
        for i in row[3:104] :
            m.append(-int(i.replace(',', '')))
        for i in row[106:] :
            f.append(int(i.replace(',', '')))

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('2020_04_Population' + name + '.Seoul_barHorizontalChart')
plt.xlabel('cnt')
plt.ylabel('age')
plt.barh(range(101), m, color = 'blue', label = 'male')
plt.barh(range(101), f, color = 'pink', label = 'female')
plt.legend()
plt.show()
```
