* 이전 시간의 도시를 입력받아 그래프를 출력하는 문제에서 '제주특별자치도'를 입력하면 오류가 난다(너무 값이 많다는) <br>
* 그 이유는 제주특별자치도를 입력하면 제주특별자치도와 그 밑의 '제주특별자치도 제주시 ~' 의 결과까지 들어가기 때문이다.
* 해결방법은 row의 data를 추가할 때 첫번째 행의 데이터만 append한 후 break문을 넣어 한 행의 데이터만 추가되도록 한다.

```python
# 입력한 동네 population
# 남_녀 분리하여 pyramid chart 그리기'

import csv

f = open('../unit_08/gender.csv')
data = csv.reader(f)

m = []
f = []

province = input('도시를 입력하세요 : ')

for row in data :
    if province in row[0] :
        for i in row[3:104] :
            m.append(-int(i.replace(',', '')))
        for i in row[106:] :
            f.append(int(i.replace(',', '')))
        break # 이게 없으면 데이터 값이 너무 많아져서 오류 발생함

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('2020_04_Population' + province + 'Chart')
plt.xlabel('cnt')
plt.ylabel('age')
plt.barh(range(101), m, color = 'blue', label = 'male')
plt.barh(range(101), f, color = 'pink', label = 'female')
plt.legend()
plt.show()
```

```python
# 레이블 추가하기
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')

size = [2442, 2312, 1031, 1233]
label = ['A형', 'B형', 'AB형', 'O형']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
explodeEffect = [0, 0, 0.1, 0] # explode : 0~1 사이, 0이면 중심점에 붙음

plt.axis('equal')
plt.pie(size, labels = label, autopct = '%.1f%%', colors = color, explode=explodeEffect) 
# %출력을 위해 %%, 
plt.legend(title = 'Blood Type')
plt.show()
```

```python
# 특정 지역에서 성별 연령별 인구 데이터 출력하기
import csv

f = open('../unit_08/gender.csv')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역 : ')

for row in data : 
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) : 
            m += int(row[i+3].replace(',', ''))
            f += int(row[i+106].replace(',', ''))
        break # '도' 로 찾기 때문
        
size.append(m)
size.append(f)
print(size)

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['male', 'female'], autopct = '%.1f%%', colors = color, startangle = 90) 
# startagle = 90 : 12시 방향부터 시작, default : 3시 방향부터 시작
plt.title(name + ' 지역의 남녀 성별 비율')
plt.show()
```

```python
# 특정 지역에서 성별 연령별 인구 데이터 출력하기
import csv

f = open('../unit_08/gender.csv')
data = csv.reader(f)

size = []

province = input('도 이름을 입력하세요 :  : ')
country = input('구 이름을 입력하세요 : ')

for row in data : 
    if province and country in row[0] :
        m = 0
        f = 0
        for i in range(101) : 
            m += int(row[i+3].replace(',', ''))
            f += int(row[i+106].replace(',', ''))
        break
        
size.append(m)
size.append(f)
print(size)

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
color = ['skyblue', 'hotpink']
plt.axis('equal')

plt.pie(size, labels=['male', 'female'], autopct = '%.1f%%', colors = color, radius = 1, startangle = 90) 
# startagle = 90 : 12시 방향부터 시작, default : 3시 방향부터 시작
plt.title(f'{province} {country} 지역의 남녀 성별 비율')
plt.show()
```
