```python
import csv
import matplotlib.pyplot as plt

f = open('../unit_02/seoul.csv')
data = csv.reader(f)
next(data) # header 제거

result = [] # 최고온도 리스트

for row in data :
    if row[-1] != '' :
        result.append(float(row[-1])) # 최고기온

plt.figure(figsize = (10,2)) # width : 10inch, height : 2inch
plt.plot(result, 'b')
plt.show()
```

```python
# 8월의 최고 기온 데이터 시각화하기
import csv
import matplotlib.pyplot as plt

f = open('../unit_02/seoul.csv')
data = csv.reader(f)
next(data) # header 제거

result = []

for row in data : 
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' : # split[1] : 월의 값
            result.append(float(row[-1]))

plt.figure(figsize = (10, 2))
plt.plot(result, 'hotpink')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('../unit_02/seoul.csv')
data = csv.reader(f)
next(data) # header 제거

result = []

for row in data : 
    if row[-1] != '' :
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '01' :
            result.append(float(row[-1]))

plt.plot(result, 'blueviolet')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('../unit_02/seoul.csv')
data = csv.reader(f)
next(data) # header 제거

high = []
low = []

year = [] # x축 값 : 년도
for i in range(1910, 2020) :
    year.append(i)

for row in data :
    if(row[-1] != '' and row[-2] != '') :
        if 1910 <= int(row[0].split('-')[0]) :
            if (row[0].split('-')[1] == '03' and row[0].split('-')[2] == '02') :
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.figure(figsize = (10, 3))

plt.title('Highest & Lowest Temperature on March 01 from 1910 to 2019')

plt.plot(high, 'hotpink', label = 'highTemp')
plt.plot(low, 'skyblue', label = 'lowTemp')
plt.legend()
plt.show()
```
