```python
import csv
import matplotlib.pyplot as plt

f = open('../unit_02/seoul.csv')
data = csv.reader(f)
next(data) # header 제거
```

```python
result = [] # 최고온도 리스트

for row in data :
    if row[-1] != '' :
        result.append(float(row[-1])) # 최고기온

plt.figure(figsize = (10,2)) # width : 10inch, height : 2inch
plt.plot(result, 'b')
plt.show()
```

```python
# 8월의 최고 기온 데이터 시각화하기
result = []

for row in data : 
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' : # split[1] : 월의 값
            result.append(float(row[-1]))

plt.figure(figsize = (10, 2))
plt.plot(result, 'hotpink')
plt.show()
```

```python
result = []

for row in data : 
    if row[-1] != '' :
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '01' :
            result.append(float(row[-1]))

plt.plot(result, 'blueviolet')
plt.show()
```
