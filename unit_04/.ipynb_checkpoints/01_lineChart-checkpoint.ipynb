#### Matplotlib : graph를 그려주는 library 
* .plot() : Plot y versus x as lines and/or markers
* .show() : Display a figure

```python
import matplotlib.pyplot as plt
plt.plot([10, 20, 30, 40]) # y value
# x value 생략시 0, 1, 2, ...
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [12, 43, 25, 15]) # ([x축], [y축])
plt.show()
```
