```python
import csv
f = open('../unit_08/gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data : 
    if name in row[0] :
        for i in range(3, 104) :
            m.append(int(row[i].replace(',', '')))
            f.append(int(row[i+103].replace(',', '')))
        break
        
import matplotlib.pyplot as plt
plt.plot(m, label = 'Male', color = 'blue')
plt.plot(f, label = 'Female', color = 'pink')
plt.legend()
plt.show()
```

### 막대그래프

```python
import csv
f = open('../unit_08/gender.csv')
data = csv.reader(f)

result = [] # 남 인구수 - 여 인구수
name = input('궁금한 동네를 입력해주세요 : ')
for row in data : 
    if name in row[0] :
        for i in range(3, 104) :
            result.append(int(row[i].replace(',', '')) - int(row[i+103].replace(',', '')))
        break
        
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(f'{name} 남성  - 여성 bar Chart')
plt.bar(range(101), result, color = 'blue', label = '남성 - 여성')
plt.legend()
plt.show()
```

### Bubble Chart

* 원의 크기로도 인구수 와 같은 특징을 나타낼 수 있다

```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40], s = [100, 200, 250, 300], c = ['red', 'blue', 'green', 'gold']) 
# s : bubble size, c : color (s, c의 갯수는 데이터 갯수와 동일해야함!)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40], s = [30, 60, 90, 120], c = range(4))
plt.colorbar() # 컬러바
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('Bubble Chart')
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40], s = [30, 60, 90, 120], c = range(4), cmap = 'jet') # cmap : color map style
plt.colorbar(orientation='horizontal', fraction=0.1) # 컬러바를 아래로
plt.show()
```

.scatter() 속성 cmap :
  * color Map, RGBA(A : alpha, 투명도)
  *  name : The name of the colormap(jet)
  * int : The number of rgb quantization levels

```python
import matplotlib.pyplot as plt
import random
x = []
y = []
size = []
for i in range(100) :
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
plt.scatter(x, y, s = size, c = size, cmap='jet', alpha=0.7) # alpha : 투명(중복되는 곳이 더 진하게 나타남)
plt.colorbar()
plt.show()
```

```python
# 특정 지역에서 성별 연령별 인구 데이터 출력하기
import csv

f = open('../unit_08/gender.csv')
data = csv.reader(f)

size = []

province = input('도 이름을 입력하세요 :  : ')
country = input('구 이름을 입력하세요 : ')

m = []
f = []

for row in data : 
    if province and country in row[0] :
        for i in range(3, 104) :
            m.append(int(row[i].replace(',', '')))
            f.append(int(row[i+103].replace(',', '')))
        break

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
color = ['skyblue', 'hotpink']
plt.axis('equal')
plt.title(f'{province} {country} 지역의 남녀 성별 비율')
plt.xlabel('남성')
plt.ylabel('여성')

plt.scatter(m, f, c = range(101), cmap = 'jet', alpha = 0.7)
plt.plot(range(max(m)), range(max(m)), 'g')
plt.colorbar()
plt.show()
```

```python
import csv
import math
f = open('../unit_08/gender.csv')
data = csv.reader(f)
m = []
f = []
size = []

province = input('도 입력 : ')
country = input('구 입력 : ')

for row in data :
    if (province and country) in row[0] :
        for i in range(3, 104) :
            m.append(int(row[i].replace(',', '')))
            f.append(int(row[i+103].replace(',', '')))
            size.append(math.sqrt(int(row[i].replace(',', '')) + int(row[i+103].replace(',', ''))))
        break
        
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(province + ' ' + country + ' 연령대별 남성, 여성')
plt.xlabel('남')
plt.ylabel('여')

plt.scatter(m, f, s = size, c = range(101), cmap = 'jet', alpha = 0.5)
plt.plot(range(max(m)), range(max(m)), 'g') # 직선(추세선)
plt.colorbar()
plt.show()
```
