```python
# 시간대 별 최대 승/하차 인원 역(새벽 4시 - 새벽3시)
import csv
f = open('../unit_12/2020_04_subwayTime.csv')
data = csv.reader(f)
next(data)
next(data)

on_mx = [0] * 24 # 승차인원
on_mx_station = [''] * 24 # 승차역
off_mx = [0] * 24 # 하차인원
off_mx_station = ['t'] * 24 #하차역
tmp = 0 # 시간

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24):
        tmp = int(str(j + 4)) # 시작이 4시부터
        if(tmp > 24) :
            tmp = tmp - 24 # ! ~ 3 시 처리
            
        tmp_passenger = row[j * 2 + 4]
        if tmp_passenger > on_mx[j] :
            on_mx[j] = tmp_passenger
            on_mx_station[j] = row[3] + ' ' + row[1] + ' ' + str(tmp) # 역, 호선, 시간
        
        tmp_passenger = row[j * 2 + 4 + 1] 
        if tmp_passenger > off_mx[j] :
            off_mx[j] = tmp_passenger
            off_mx_station[j] = row[3] + ' ' + row[1] + ' ' + str(tmp) 


import matplotlib.pyplot as plt
# 승차
plt.figure(figsize = (10, 3))
plt.rc('font', family = 'Malgun Gothic')
plt.title('2020년 4월 시간대별로 최대 승차인원')

plt.bar(range(24), on_mx, color = 'hotpink', label = '승차')
plt.xticks(range(24), on_mx_station, rotation = 90)
plt.savefig('./images/' + '2020년 4월 시간대별로 최대 승차인원' + '.png', bbox_inches='tight')

# 하차
plt.figure(figsize = (10, 3))
plt.rc('font', family = 'Malgun Gothic')
plt.title('2020년 4월 시간대별로 최대 하차인원')

plt.bar(range(24), off_mx, color = 'skyblue', label = '하차')
plt.xticks(range(24), off_mx_station, rotation = 90)
plt.savefig('./images/' + '2020년 4월 시간대별로 최대 하차인원' + '.png', bbox_inches='tight')
plt.show()
```
