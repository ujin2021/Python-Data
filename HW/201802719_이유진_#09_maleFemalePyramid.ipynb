```python
import csv

f = open('../unit_08/gender.csv')
data = csv.reader(f)

m = []
f = []

name = input('구와 읍면동 단위로 입력하세요 : ') # 처인구 모현

for row in data :
    if name in row[0] :
        for i in row[3:104] : # 남자 0~100세이상
            m.append(-int(i.replace(',', '')))
        for i in row[106:] : # 여자 0~100세 이상
            f.append(int(i.replace(',', '')))

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # 음수값 오류처리

plt.title('2020_04_Population ' + name + ' Pyramid Chart')

plt.xlabel('cnt')
plt.ylabel('age')

plt.barh(range(101), m, color = 'blue', label = 'male')
plt.barh(range(101), f, color = 'pink', label = 'female')

plt.legend()
plt.show()
```
