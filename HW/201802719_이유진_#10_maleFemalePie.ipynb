```python
# 특정 지역에서 성별 연령별 인구 데이터 출력하기
import csv

f = open('../unit_08/gender.csv')
data = csv.reader(f)

size = []

province = input('도 이름을 입력하세요 :  : ')
country = input('구 이름을 입력하세요 : ')

for row in data : 
    if province and country in row[0] :
        m = 0
        f = 0
        for i in range(101) : 
            m += int(row[i+3].replace(',', ''))
            f += int(row[i+106].replace(',', ''))
        break
        
size.append(m)
size.append(f)

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
label = ['male', 'female']
color = ['skyblue', 'hotpink']
explodeEffect = [0, 0.1]
plt.axis('equal')

plt.pie(size, labels=label, autopct='%.2f%%', colors=color, startangle=90, explode=explodeEffect, shadow=True)
plt.title(f'{province} {country} 지역의 남녀 성별 비율')
plt.legend(title = '성별')
plt.show()
```
